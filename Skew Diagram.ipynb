```python
dna_string = input()
cnt_g = 0
cnt_c = 0
points_x = [0]
points_y = [0]
```

```python
for i, c in enumerate(dna_string):
    if(c == "G" or c == "g"):
        cnt_g = cnt_g + 1
    elif(c == "C" or c == "c"):
        cnt_c = cnt_c + 1
    points_y.insert(i+1, cnt_g - cnt_c)
    points_x.insert(i+1, i+1)
    print(i, cnt_g, cnt_c)
#print(points_x)
#print(points_y)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(1,15))
plt.plot(points_x, points_y)
plt.ylabel('G-C')
plt.xlabel('Index')
plt.show()
```
