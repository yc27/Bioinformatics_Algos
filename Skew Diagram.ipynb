```python
dna_string = input()
cnt_g = 0
cnt_c = 0
points_x = [0]
points_y = [0]
```

```python
for i, c in enumerate(dna_string):
    if(c == "G"):
        cnt_g += 1
    elif(c == "C"):
        cnt_c += 1
    points_x.insert(i+1, cnt_g - cnt_c)
    points_y.insert(i+1, i+1)
```

```python
import matplotlib.pyplot as plt
plt.plot(points_y, points_x)
plt.ylabel('G-C')
plt.xlabel('Index')
plt.show()
```
